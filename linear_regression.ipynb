```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1, 1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 78, 55, 67, 57, 4, 12]).reshape(-1, 1)

display(time_studied)
```

Creating the model

```python
model = LinearRegression()
model.fit(time_studied, scores)
```

To visualize in graphic

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(time_studied, scores, c="purple", linewidths=5)
ax.set_ylim(0,100) # The max y values
plt.show()
```

Using the linear regression

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(time_studied, scores)

ax.plot(np.linspace(0, 70, 100).reshape(-1, 1),
        model.predict(np.linspace(0, 70, 100).reshape(-1, 1)),
        'r',
        linewidth=3,
        label="Point by Time")  # To predict

ax.set_ylim(0,100) # The max y values
ax.set_ylabel("Time of Study")
ax.set_xlabel("Score")
plt.show()
```

```python
hours_feature = 56
predicted_scored = model.predict(np.array([hours_feature]).reshape(-1, 1))# To predict which will be my score if i studied xxx hours

print(f"If you study for {hours_feature} hours, you score propably will be : [{predicted_scored[0][0]:.0f}] ")
```
